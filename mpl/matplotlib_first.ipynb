```python
x=[-3,-2,-1,0,1,2,3,4,5]
y=[9,4,1,0,1,4,9,16,25]
ppt.plot(x,y)
```

```python
x=[]
y=[]
for i in range (-7,8):
    x.append(i)
    y.append(i**3)
ppt.plot(x,y)
ppt.title("cubes")
```

```python
x=[]
y=[]
for i in range (-7,8):
    x.append(i)
    y.append(i**3)
ppt.bar(x,y)
ppt.title("cubes")
```
